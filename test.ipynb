```python
from ultralytics import YOLO
import cv2

# Load trained model
model = YOLO("runs/detect/Normal_Compressed/weights/best.pt")

# Open video file (or use 0 for webcam)
video_path = r"C:\Users\JAYESH\Downloads\vedio 2.mp4"  # Change this to your test video path
cap = cv2.VideoCapture(video_path)

# Get video properties
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Output video settings
out = cv2.VideoWriter("output.mp4", cv2.VideoWriter_fourcc(*"mp4v"), fps, (frame_width, frame_height))

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # Run YOLOv8 inference on frame
    results = model(frame, conf=0.5)

    # Draw bounding boxes on frame
    for r in results:
        frame = r.plot()

    # Show frame
    cv2.imshow("Weapon Detection", frame)
    out.write(frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):  # Press 'q' to exit
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
