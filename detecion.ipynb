```python
from ultralytics import YOLO

# Load the model
model = YOLO("runs/detect/Normal_Compressed/weights/best.pt")  

# Run evaluation
metrics = model.val(data="C:/Users/JAYESH/Desktop/Weapons-and-Knives-Detector-with-YOLOv8/data.yaml")

# Print evaluation metrics correctly
print(f"mAP50: {metrics.box.map50:.4f}")  # Mean Average Precision at 50% IoU
print(f"mAP50-95: {metrics.box.map:.4f}")  # Mean Average Precision at 50-95% IoU

# Check if precision, recall, and F1 score exist in the metrics object
if hasattr(metrics.box, 'p') and hasattr(metrics.box, 'r') and hasattr(metrics.box, 'f1'):
    precision = metrics.box.p  # List of precision values for each class
    recall = metrics.box.r  # List of recall values for each class
    f1_score = metrics.box.f1  # List of F1 scores for each class

    # Print average precision, recall, and F1-score across all classes
    print(f"Precision: {sum(precision) / len(precision):.4f}")
    print(f"Recall: {sum(recall) / len(recall):.4f}")
    print(f"F1 Score: {sum(f1_score) / len(f1_score):.4f}")
else:
    print("Precision, recall, and F1-score are not available in the metrics object.")
```
